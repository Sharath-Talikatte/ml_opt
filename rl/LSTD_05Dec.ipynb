```python
import numpy as np
import matplotlib.pyplot as plt
import tqdm
import math
import sklearn
import scipy
from scipy.linalg import lstsq
from sklearn.metrics import mean_squared_error
import os
import random
import timeit

np.set_printoptions(suppress=True)

def env(current_state, action):
    if action == '1':
        action_taken = -1
    elif action == '2':
        action_taken = -2
    else:
        print('Error')
    if current_state >= 2:
        next_state = current_state + action_taken
        single_stage_reward = -3
    elif current_state == 1:
        next_state = 0
        single_stage_reward = -2
    else:
        print('Error')
    return next_state, single_stage_reward

def generate_random_number():
    random_bytes = os.urandom(1)
    random_number = int.from_bytes(random_bytes, byteorder='big')
    rand = float(random_number)
    return rand / 255

def take_action_as_per_pi(pi):
    ##test = generate_random_number()
    test = random.random()
    if test <= pi:
        action = '1'
    else:
        action = '2'
    return action

def get_step_size(a0, n0, n):
    alpha_n = a0 * (n0 + 1) / (n0 + n)
    return alpha_n

def rms_error(weight_vector):
    ideal_values = np.matmul(feature_vector, true_weight_vector)
    current_values = np.matmul(feature_vector, weight_vector)
    mse = mean_squared_error(ideal_values, current_values)
    return mse

def td_lambda(a0, n0, num_episodes, lambda_value, gamma=1):
    weight_vectors = []
    for trial in range(NUM_TRIALS_FOR_AVERAGING):
        weight_vector = np.zeros(4, dtype=np.float32)
        for n in range(num_episodes):
            sigma = np.zeros_like(weight_vector)
            ##current_state = random.randint(1,12)
            current_state = 12
            z = feature_vector[current_state]
            while current_state > 0:
                current_action = take_action_as_per_pi(pi)
                next_state, single_stage_reward = env(current_state, current_action)
                sigma = sigma + z * (single_stage_reward + np.dot(np.subtract(gamma * feature_vector[next_state],feature_vector[current_state]), weight_vector))
                z = np.add(lambda_value * z, feature_vector[next_state])
                current_state = next_state
            weight_vector = weight_vector + get_step_size(a0, n0, n)* sigma
        weight_vectors.append(weight_vector)
    final_weight_vector = np.average(weight_vectors, axis=0)
    rms = rms_error(final_weight_vector)
    return rms,final_weight_vector

def regularize_matrix(matrix, factor):
    d = matrix.shape[0]
    output_matrix = matrix.copy()
    for i in range(d):
        for j in range(d):
            if i == j:
                output_matrix[i,j] = matrix[i,j] + factor
    return output_matrix

def find_inverse(A, threshold):
    A_copy = A.copy()
    U, Sigma, Vt = np.linalg.svd(A_copy)
    Sigma_pseudo_inv = np.zeros_like(A_copy, dtype=float)
    d = Sigma_pseudo_inv.shape[0]
    for i in range(d):
        diag_entry = Sigma[i]
        if abs(diag_entry) > threshold:
            Sigma_pseudo_inv[i,i] = 1 / Sigma[i]
        else:
            Sigma_pseudo_inv[i,i] = 0
    A_pseudo_inv = Vt.T @ Sigma_pseudo_inv @ U.T
    return A_pseudo_inv

def lstd_lambda(num_episodes, lambda_value , gamma=1):
    weight_vectors = []
    for trial in range(NUM_TRIALS_FOR_AVERAGING):
        z = np.zeros(4)
        A = np.zeros((4,4), dtype = np.float32)
        rms_error_list = []
        b = np.zeros(4)
        for n in range(num_episodes):
            weight_vector = np.zeros(4, dtype=np.float32)
            current_state = 12
            z = feature_vector[current_state]
            while current_state > 0:
                current_action = take_action_as_per_pi(pi)
                next_state, single_stage_reward = env(current_state, current_action)
                A = A + np.outer(z, feature_vector[current_state] - gamma * feature_vector[next_state])
                b = b + z * single_stage_reward
                z = gamma * lambda_value * z + feature_vector[next_state]
                current_state = next_state
            weight_vector = np.dot(np.linalg.inv(A), b)
        weight_vectors.append(weight_vector)
    final_weight_vector = np.average(weight_vectors, axis=0)
    rms = rms_error(final_weight_vector)
    return rms,final_weight_vector

def rlstd_lambda(num_episodes, lambda_value, gamma=1):
    weight_vectors = []
    for trial in range(NUM_TRIALS_FOR_AVERAGING):
        weight_vector = np.zeros(4, dtype=np.float32)
        Ainv = np.eye(feature_vector.shape[1]) * 1/10e-10
        rms_error_list = []
        A = np.eye(4, dtype = np.float32)*0.01
        b = np.zeros_like(weight_vector)
        for n in range(num_episodes):
            current_state = 12
            #current_state = random.randint(2,12)
            phi_t = feature_vector[current_state].reshape(-1,1)
            z = phi_t
            while current_state > 0:
                current_action = take_action_as_per_pi(pi)
                next_state, single_stage_reward = env(current_state, current_action)
                phi_tp = feature_vector[next_state].reshape(-1,1)
                u = z
                v = phi_t -  gamma * phi_tp
                Ainv = Ainv - np.dot(Ainv, np.dot(np.dot(u, v.T), Ainv))/(1 + np.dot(v.T, np.dot(Ainv, u)))
                b = b + z.reshape(-1) * single_stage_reward  #b = b + eligibility * reward
                z = gamma * lambda_value * z + phi_tp #Set $Z_t+1 = \lambda * Z_t + phi(x_t+1)
                phi_t = phi_tp
                current_state = next_state
            weight_vector = np.dot(Ainv,b)
            weight_vectors.append(weight_vector)
    final_weight_vector = np.average(weight_vectors, axis=0)
    rms = rms_error(final_weight_vector)
    return rms,final_weight_vector

# Feature Vector Definition - PHI
feature_vector = np.array([[0,0,0,0],
                  [0,0,0.25,0.75],
                  [0,0,0.5,0.5],
                  [0,0,0.75,0.25],
                  [0,0,1,0],
                  [0,0.25,0.75,0],
                  [0,0.5,0.5,0],
                  [0,0.75,0.25,0],
                  [0,1,0,0],
                  [0.25,0.75,0,0],
                  [0.5,0.5,0,0],
                  [0.75,0.25,0,0],
                  [1,0,0,0]])

true_weight_vector = np.array([-24,-16,-8,0])

NUM_EPISODES_LIST = []

for i in np.linspace(1,100,10, dtype=int):
    NUM_EPISODES_LIST.append(i)
for i in np.linspace(100,1000,10, dtype=int):
    NUM_EPISODES_LIST.append(i)
for i in np.linspace(1000,10000,20, dtype=int):
    NUM_EPISODES_LIST.append(i)

pi = 0.5
lambda_list = [0.00, 0.25, 0.40, 0.50, 0.75, 1.00]

a0_list = [0.1,0.01]
n0_list = [100, 1000, 1000000]

# Exp to show sensitivity of feature vector range
change_factor = 20
run_exp = False
if run_exp:
  feature_vector = change_factor * feature_vector
  true_weight_vector = true_weight_vector / change_factor

NUM_TRIALS_FOR_AVERAGING = 10
td_plot_data = np.zeros((len(lambda_list), len(a0_list), len(n0_list), len(NUM_EPISODES_LIST)))
lstd_plot_data = np.zeros((len(lambda_list),len(NUM_EPISODES_LIST)))
rlstd_plot_data = np.zeros((len(lambda_list),len(NUM_EPISODES_LIST)))
total_runs = td_plot_data.size

def run_data_gen():
    ## TD
    try:
        for lm_idx, lm_val in enumerate(lambda_list):
            for a0_count, a0 in enumerate(a0_list):
                for n0_count, n0 in enumerate(n0_list):
                    for num_episodes_count, num_episodes in tqdm.tqdm(enumerate(NUM_EPISODES_LIST),total=len(NUM_EPISODES_LIST),desc="TD(lambda)", colour= "GREEN"):
                        td_plot_data[lm_idx, a0_count, n0_count, num_episodes_count],weights = td_lambda(a0, n0, num_episodes,lm_val)
                    print(f'a0: {a0} | n0: {n0} | Lambda: {lm_val} | Weights: {weights}\n')
        print('\n')
    except Exception as e:
        print("Failed to run TD")
    
    # LSTD
    for lm_idx, lm_val in enumerate(lambda_list):
        for num_episodes_count, num_episodes in tqdm.tqdm(enumerate(NUM_EPISODES_LIST),total=len(NUM_EPISODES_LIST),desc="LSTD(lambda)",colour= "GREEN"):
            lstd_plot_data[lm_idx,num_episodes_count],weights = lstd_lambda(num_episodes,lm_val)
        print(f'Lambda: {lm_val} | Weights: {weights}\n')
    print('\n')

    ## RLSTD
    for lm_idx, lm_val in enumerate(lambda_list):
        for num_episodes_count, num_episodes in tqdm.tqdm(enumerate(NUM_EPISODES_LIST),total=len(NUM_EPISODES_LIST),desc="RLSTD(lambda)",colour= "GREEN"):
            rlstd_plot_data[lm_idx,num_episodes_count],weights = rlstd_lambda(num_episodes,lm_val)
        print(f'Lambda: {lm_val} | Weights: {weights}\n')
    rlstd_done = timeit.default_timer()

    return td_plot_data, lstd_plot_data, rlstd_plot_data

def plot_data(td_plot_data, lstd_plot_data, rlstd_plot_data):
    ## Plot for TD
    plt.figure(figsize=(10, 4))
    plt.xscale('log')
    plt.xlabel('episode')
    plt.ylabel('TD - RMS error over all states')
    for a0_count, a0 in enumerate(a0_list):
        for n0_count, n0 in enumerate(n0_list):
            X = NUM_EPISODES_LIST
            Y = td_plot_data[0, a0_count, n0_count]
            Y_clip = np.clip(Y,0,1)
            legend = 'TD_lambda=0.4_' + str(a0) + '_' + str(n0)
            plt.plot(X, Y_clip, label=legend)
    plt.legend(fontsize=8)
    plt.grid()
    plt.show()

    ## Plot for LSTD
    plt.figure(figsize=(10, 4))
    plt.xscale('log')
    plt.xlabel('episode')
    plt.ylabel('LSTD - RMS error over all states')
    for lm_idx, lm_val in enumerate(lambda_list):
        X = NUM_EPISODES_LIST
        Y = lstd_plot_data[lm_idx]
        Y_clip = np.clip(Y,0,20)
        legend = 'LSTD_lambda_' + str(lm_val)
        plt.plot(X, Y_clip, label=legend)
    plt.legend(fontsize=8)
    plt.grid()
    plt.show()

    ## Plot for RLSTD
    plt.figure(figsize=(10, 4))
    plt.xscale('log')
    plt.xlabel('episode')
    plt.ylabel('RLSTD - RMS error over all states')
    for lm_idx, lm_val in enumerate(lambda_list):
        X = NUM_EPISODES_LIST
        Y = rlstd_plot_data[lm_idx]
        Y_clip = np.clip(Y,0,20)
        legend = 'RLSTD_lambda_' + str(lm_val)
        plt.plot(X, Y_clip, label=legend)
    plt.legend(fontsize=8)
    plt.grid()
    plt.show()
    return

def plot_data_ranges(td_plot_data, lstd_plot_data):

    plt.figure(figsize=(10, 4))
    plt.yscale('log')
    plt.xlabel('lambda')
    plt.ylabel('RMS error over all states')

    pos = np.linspace(-0.03,0.03,td_plot_data.shape[1]*td_plot_data.shape[2] + 1)
    plt_style = ['--ko','--ro','--bo','--go','--mo','--co','--yo']
    i = 0
    a = np.array(NUM_EPISODES_LIST)
    for a0_count, a0 in enumerate(a0_list):
        for n0_count, n0 in enumerate(n0_list):
            X = lambda_list
            Y = td_plot_data[:,a0_count,n0_count,[np.where(a==100)[0][0],np.where(a==1000)[0][0]]].reshape(len(lambda_list),-1)
            for idx,(xe, ye) in enumerate(zip(X, Y)):
                if idx == 0:
                    legend = 'TD_lambda_' + str(a0) + '_' + str(n0)
                    plt.plot([xe + pos[i]] * len(ye), ye, plt_style[i%7] , label=legend)
                else:
                    plt.plot([xe + pos[i]] * len(ye), ye, plt_style[i%7])
            i += 1
    
    Y = lstd_plot_data[:,[np.where(a==100)[0][0],np.where(a==1000)[0][0]]].reshape(len(lambda_list),-1)
    for idx, (xe, ye) in enumerate(zip(X, Y)):
        if idx == 0:
            legend = 'LSTD_lambda'
            plt.plot([xe + pos[-1]] * len(ye), ye, plt_style[i%7], label=legend)
        else:
            plt.plot([xe + pos[i]] * len(ye), ye, plt_style[i%7])
    plt.legend(fontsize=8, loc='center left', bbox_to_anchor=(1, 0.5))
    plt.grid()
    plt.show()
    return 0


td_plot_data, lstd_plot_data, rlstd_plot_data = run_data_gen()
plot_data(td_plot_data, lstd_plot_data, rlstd_plot_data)
plot_data_ranges(td_plot_data, lstd_plot_data)
```
